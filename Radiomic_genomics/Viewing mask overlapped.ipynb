# Overlay image and its mask
The aim of this report is to prove that after setting the exact geomatrucal parameters to the mask, both the image and the mask perfectly overlay. By plotting it, we can be sure that PyRadiomics' results are reliable.

```python
from nilearn import plotting
```

Below, be can see two groups of images and masks. The first group contains a wrong mask with only a few geometrical parameters matching between the mask and the image. The second group contains a mask with the exact geometrical parameters than the original image.

```python
# Imagenes con parametros geometricos diferentes
path_to_corrected_mask = '/home/carlos/ISGlobal/patients_nii/rstudio-export/023140000001/023140000001_INSP_STD_L1_ECLIPSE_corrected_geometry_mask.nii.gz'
path_to_original_image = '/home/carlos/ISGlobal/patients_nii/rstudio-export/023140000001/023140000001_INSP_STD_L1_ECLIPSE.nii.gz'
path_to_original_mask = '/home/carlos/ISGlobal/patients_nii/rstudio-export/023140000001/023140000001_INSP_STD_L1_ECLIPSE_mask.nii.gz'

# Imagenes con la geometria correcta
image_path_test = '/home/carlos/ISGlobal/prueba_extraccion_PyRad/023140000001_INSP_STD_L1_ECLIPSE.nii.gz'
mask_path_test = '/home/carlos/ISGlobal/prueba_extraccion_PyRad/023140000001_INSP_STD_L1_ECLIPSE_mask.nii.gz'
path_corrected_geometry = '/home/carlos/ISGlobal/prueba_extraccion_PyRad/023140000001_INSP_STD_L1_ECLIPSE_corrected_geometry_mask.nii.gz'
```

Plotting the not perfectly matching mask and image:

```python
display = plotting.plot_anat(path_to_original_image)
display.add_contours(path_to_corrected_mask,
                     linewidths=2.5, levels=[.5],
                     cmap=plotting.cm.purple_green)
plotting.show()
```

```python
display = plotting.plot_anat(path_to_original_image)
display.add_overlay(path_to_corrected_mask,
                     cmap=plotting.cm.purple_green)
plotting.show()
```

We can observe that if the geometrical parameters are not exactly the same between the image and the mask there is no overlay. Therefore, if the parameters are the same, both files will overlay.

```python
display = plotting.plot_anat(image_path_test)
display.add_contours(path_corrected_geometry,
                     linewidths=2.5, levels=[.5],
                     cmap=plotting.cm.blue_transparent)
plotting.show()
```

```python
display = plotting.plot_anat(image_path_test)
display.add_overlay(path_corrected_geometry,
                     cmap=plotting.cm.purple_green)
plotting.show()
```

To exctract the radiomic features with PyRadiomics we used the files with the matching geometrical parameters. Therefore, we can affrim that the results are reliable.
